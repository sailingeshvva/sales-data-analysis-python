```python
for i in ["quantity_ordered","order_id","price_each"]:
    df[i]=pd.to_numeric(df[i],errors="coerce")
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format="%m/%d/%y %H:%M",errors="coerce")
```

```python
df.dtypes
```

```python
# Adding Additional Fields

df["city"] = df["purchase_address"].apply(lambda x: x.split(",")[1].strip() if isinstance(x, str) and "," in x else None)


df["Year"] = df["order_date"].dt.year
df["Month"] = df["order_date"].dt.month
df["Week"] = df["order_date"].dt.isocalendar().week
df["Weekday"] = df["order_date"].dt.day_name()

df["Hour"] = df["order_date"].dt.hour

df["Total_Sale_Price"] = df["price_each"] * df["quantity_ordered"]
df.head()
```

```python
# 5. Heatmap - correlation matrix of numerical fectures
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap of numerical Features")
plt.tight_layout()
plt.show()
```

```python
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

plt.figure(figsize=(10, 4))
plt.title("Sales By Week Day")
grouped_weekdays = df.groupby("Weekday")["order_id"].nunique().reindex(weekday_order)
grouped_weekdays.plot(kind="line")
plt.grid(True)
plt.xlabel("Weekday")
plt.ylabel("Order by Weekday")
plt.savefig("chart.png") 
plt.show()
```

# Sales by Hour (Scatter Plot)

Most orders happen around mid-morning and evening, showing peak shopping hours when customers are most active.

```python
orders_per_hour = df.groupby("Hour")["order_id"].nunique().reset_index()

orders_per_hour.columns = ["Hour", "unique_orders"]

plt.figure(figsize=(18, 6))
plt.plot(orders_per_hour["Hour"], orders_per_hour["unique_orders"], marker="o")
plt.title("Unique Orders by Hour of Day")
plt.xlabel("Hour of Day")
plt.xticks(range(0, 24), labels=[f"{h:02d}:00" for h in range(0, 24)])
plt.ylabel("Number of Unique Orders")
plt.grid(True)
plt.show()
```

#  Sales by City (Bar Chart)

Some cities generate much higher sales, indicating where most of our customer base is located.

```python
sales_by_region = df.groupby("city")["Total_Sale_Price"].sum().reset_index()
sales_by_region = sales_by_region.sort_values("Total_Sale_Price", ascending=False)

plt.figure(figsize=(12,4))
plt. title("Sales By City")
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.ylabel("Total Sales (Dollars)")
sns.barplot(data=sales_by_region, x="city", y="Total_Sale_Price")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="quantity_ordered", y="product", data=df)
plt.title("Quantity Ordered For Each Product")
plt.xlabel("Quantity Ordered")
plt.ylabel("Product")
plt.show()
```

```python
top10 = df.iloc[9:, :].reset_index()

plt.figure(figsize=(16, 8))
sns.barplot(data=top10, x='product', y='Total_Sale_Price', palette='magma', legend=False)
plt.title('Top 10 selling products in terms of revenue in USA')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
